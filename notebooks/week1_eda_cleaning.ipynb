```python
import pandas as pd
import numpy as np

# 1. Load the data
# Make sure your data folder is in your .gitignore before running this!
df = pd.read_csv('../data/online-retail-dataset.csv', encoding='ISO-8859-1')

print("Initial Shape:", df.shape)

# 2. Week 1 Task: Filter out missing CustomerIDs and failed/refunded orders
# Clean missing user IDs
df = df.dropna(subset=['CustomerID']) 

# Filter out refunds (in this dataset, invoice numbers starting with 'C' are cancellations)
df = df[df['Quantity'] > 0]
df = df[df['UnitPrice'] > 0]

# 3. Standardize Timestamp formats
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# 4. Week 1 Task: Calculate "Cohort Month" (Month of first transaction)
# Extract Year-Month from Invoice Date
df['InvoiceMonth'] = df['InvoiceDate'].dt.to_period('M')

# Find the first purchase month for each unique customer
df['CohortMonth'] = df.groupby('CustomerID')['InvoiceMonth'].transform('min')

print("Cleaned Shape:", df.shape)
print(df[['CustomerID', 'InvoiceMonth', 'CohortMonth']].head())
```
