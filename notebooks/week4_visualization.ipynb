```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# STEP 1: Load the percentage matrix saved in Week 2
try:
    matrix_pct = pd.read_pickle('../data/retention_matrix_pct.pkl')
    print("Successfully loaded Week 2 retention matrix data!")
except FileNotFoundError:
    print("Error: Could not find 'retention_matrix_pct.pkl'. Make sure you ran your Week 2 notebook successfully!")

# STEP 2: The Heatmap Generation Function 
def generate_heatmap(matrix_pct):
    # Set up the style and size of the plot 
    plt.figure(figsize=(14, 10))
    
    # Title styling 
    plt.title('Cohort Retention Rates (%) - User Stickiness', fontsize=16, fontweight='bold')
    
    # Build seaborn heatmap with nice formatting 
    sns.heatmap(data=matrix_pct, 
                annot=True, 
                fmt='.1%', 
                cmap='BuGn', 
                cbar_kws={'label': 'Retention Rate'})
    
    plt.xlabel('Cohort Index (Months Elapsed)')
    plt.ylabel('Cohort Join Month')
    
    # Save chart cleanly to your notebook folder [cite: 81]
    plt.savefig('cohort_retention_heatmap.png', dpi=300, bbox_inches='tight')
    
    # STEP 3: FORCE THE CHART TO DISPLAY ON SCREEN
    plt.show()


# STEP 4: RUN THE FUNCTION!

generate_heatmap(matrix_pct)
```
